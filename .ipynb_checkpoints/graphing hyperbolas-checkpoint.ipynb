```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = .5
h = 1
k = 2

axlim = 2*np.max((a,b)) + ((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

# create the function
hyperbola = (X-h)**2/a**2 - (Y-k)**2/b**2 - 1

plt.contour(X,Y,hyperbola,2)
plt.plot(h,k,'go')

plt.plot([-axlim,axlim],[0,0])
plt.plot([0,0],[-axlim,axlim],'r--',color=[.8,.8,.8])
plt.gca().set_aspect('equal')
plt.show()
```

```python
# exercise.independent
a = list(range(11))
b = a
h = 0
k = 0

axlim = 2*np.max((a,b)) + ((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

for i in a:
    hyperbola = (X-h)**2/a[i]**2 - (Y-k)**2/b[i]**2 - 1
    plt.contour(X,Y,hyperbola)
    
plt.gca().set_aspect('equal')
# 
plt.show()
```
