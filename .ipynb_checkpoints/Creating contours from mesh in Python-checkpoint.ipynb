```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display,Math
import sympy as sym
```

```python
X,Y = np.meshgrid(range(0,10),range(0,15)) #upper case letters usually indicate matrices

plt.subplot(121)
plt.pcolormesh(X,edgecolor='k',linewidth=.1)

plt.subplot(122)
plt.pcolormesh(Y,edgecolor='k',linewidth=.1)

plt.show()
```

```python
x = np.linspace(-np.pi,2*np.pi,40)
y = np.linspace(0,4*np.pi,72)

X,Y = np.meshgrid(x,y) #only good to use i you dont have too high of resolution

Fxy = np.cos(X) + np.sin(Y)

plt.imshow(Fxy)
plt.show()
```

```python

x = np.linspace(-2,2,100)
X,Y = np.meshgrid(x,x)
s = 2

gauss2d = np.exp( -(X**2 + Y**2)/s)

plt.imshow(gauss2d)
plt.axis('off')
plt.show()
```
