```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1 # 1/4y
h = -10
k = -2
n = 100

# x-axis point to evaluate the function
x = np.linspace(h-5,h+5,n)

# create the funtion
y = a*(x-h)**2 + k

plt.plot(y,x)
plt.grid()

plt.show()
```

```python
# exerise.independent
a = 1 # 1/4y
h = -10
k = -2
n = 100
p = 1

# x-axis point to evaluate the function
x = np.linspace(h-5,h+5,n)

# create the funtion
y = a*(x-h)**2 + k

# plot parabola
plt.plot(x,y)

# plot focus
plt.plot([h], [k+p],'ro')

# plot directrix
plt.plot([x],[k-p],'yo')

# plot vertex
plt.plot(h,k,'go')


plt.legend()
plt.show()
```

```python
# exercise.instructed
a = 1 # 1/4p = a, 1 = a4p, , 1/4a
p = 1/(4*a)
h = -10
k = -2
n = 100

# x-axis point to evaluate the function
x = np.linspace(h-2,h+2,n)

# create the function
y = a*(x-h)**2 + k

# plot the parabola
plt.plot(x,y,label='Parabola')

# plot the vertex
plt.plot(h,k,'ro',label='Vertex')

# plot the focus
plt.plot(h,k+p,'go',label='Focus')

# plot the directrix
d = k-p
plt.plot(x[[0,-1]],[d,d],label='Directrix')

plt.legend()
plt.grid()
plt.show()
```
